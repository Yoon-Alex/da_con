## 유동인구 데이터 전처리

```python
direc = "./KT_data/"

fpopl_columns = pd.read_csv(direc + "fpopl.csv", nrows = 1) # 행정동 유동인구
dtype = dict(zip(list(fpopl_columns.columns), ['object'] * 5 + ['int']))

fpopl = pd.read_csv(direc + "fpopl.csv", dtype = dtype) # 행정동 유동인구
fpopl.head() # 24시간 구분, 성별, 연령, 시군구, 유동인구
```

```python
# 약 5천만 row -> too heavy 
fpopl.shape 
```

```python
data_by_rg = fpopl.groupby(['base_ymd', 'adstrd_code'], as_index=False).agg({'popltn_cascnt': 'sum'})
```

```python
data_by_rg.shape
```

```python
pop_daily = data_by_rg.groupby('base_ymd', as_index=False).agg({'popltn_cascnt': 'sum'})
```

```python
pop_daily.head()
```

```python
pop_daily.base_ymd = pop_daily.base_ymd.astype("datetime64")
```

```python
pop_daily.set_index('base_ymd', inplace = True)
```

```python
data_by_rg.head()
```

```python
df = data_by_rg
```

```python
df.base_ymd = df.base_ymd.astype("datetime64")
```

```python
df.adstrd_code = df.adstrd_code.astype("str")
```

```python
df.set_index("base_ymd", inplace = True)
```

```python
adstrd_master = adstrd_master.astype('str')
```

```python
df = df.reset_index().merge(adstrd_master, how = 'left', on='adstrd_code').set_index("base_ymd")
```

```python
high = list(df.groupby("adstrd_code").popltn_cascnt.mean().sort_values(ascending = False)[:10].index)
low = list(df.groupby("adstrd_code").popltn_cascnt.mean().sort_values(ascending = False)[-10:].index)
```

```python
high + low
```

```python
df = df.query("adstrd_code in @high or adstrd_code in @low")
```

```python
# 지역 by 지역으로 유동인구가 다름. 직장, 상업지역에 유동인구가 모일 것으로 예측
df.query("adstrd_code in @high or adstrd_code in @low") \
  .pivot_table(index = df.index, columns = 'adstrd_nm', values = 'popltn_cascnt') \
  .rolling(window = 7).mean() \
  .iplot("line")
```

### 연령별 유동인구에 대한 분석 진행 
### 코로나 확진 환자 수에 따른 분석 진행

```python
# 평균 유동인구 수 비교
high_popu = list(data_by_rg.groupby("adstrd_code").popltn_cascnt.mean().sort_values(ascending = False)[:20].index)
```

```python
fpopl.query("adstrd_code in @high_popu") \
     .groupby(['adstrd_code','agrde_se_code']) \
     .agg({'popltn_cascnt' : 'sum'}).reset_index() \
     .merge(adstrd_master, how = 'left', on='adstrd_code') \
     .pivot(index = 'agrde_se_code', columns = 'adstrd_nm', values = 'popltn_cascnt') \
     .iplot("line")
```

```python
result = fpopl.query("adstrd_code in @high_popu and \
             agrde_se_code in ['age_20','age_25','age_30','age_35']") \
         .groupby(['base_ymd','adstrd_code']) \
         .agg({'popltn_cascnt' : 'sum'}).reset_index() \
         .merge(adstrd_master, how = 'left', on='adstrd_code')
```

```python
result.pivot_table(index = pd.to_datetime(result.base_ymd), columns = 'adstrd_nm', values = 'popltn_cascnt')  \
      .rolling(7).mean() \
      .iplot("line")    
```
