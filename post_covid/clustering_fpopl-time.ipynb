## 유동인구 데이터 전처리 (클러스터링 지역 분석)

```python
direc = "./KT_data/"

fpopl_columns = pd.read_csv(direc + "fpopl.csv", nrows = 1) # 행정동 유동인구
dtype = dict(zip(list(fpopl_columns.columns), ['object'] * 5 + ['int']))

fpopl = pd.read_csv(direc + "fpopl.csv", dtype = dtype) # 행정동 유동인구
fpopl.head() # 24시간 구분, 성별, 연령, 시군구, 유동인구
```

```python
# 약 5천만 row -> too heavy 
fpopl.shape 
```

```python
rep = dict(zip(list(fpopl.agrde_se_code.unique()) , ['age_0019'] * 3 + ['age_2039'] * 4 + ['age_4059'] * 4 + ['age_60~'] * 3))
```

```python
fpopl.replace(rep, inplace = True)
```

```python
df = fpopl.groupby(['base_ymd', 'adstrd_code', 'tmzon_se_code'], as_index=False).agg({'popltn_cascnt': 'sum'})
```

```python
df.base_ymd = df.base_ymd.astype("datetime64")
```

```python
df.tmzon_se_code = df.tmzon_se_code.str.zfill(2)
```

```python
# 지역정보 매핑 
adstrd_master = pd.read_csv('./KT_data/adstrd_master.csv', dtype = {'adstrd_code' : 'int'}) 
```

```python
adstrd_master.adstrd_code = adstrd_master.adstrd_code.astype('str')
```

```python
adstrd_master
```

```python
df = df.merge(adstrd_master, how = 'left', on = 'adstrd_code')
```

```python
df
```

```python
# 공휴일 정보
holi = pd.read_csv("../../13. 공공데이터 API 크롤링/holiday.csv", dtype = 'str')
holi.head()
```

```python
holi['date'] = pd.to_datetime(holi['0'])
```

```python
df.query("base_ymd.dt.weekday >= 5")
```

```python
def make_holi(date):
    if date.weekday() >= 5:
        return 'Y'
    else :
        return 'N'
```

```python
df['wn_yn'] = df.base_ymd.apply(make_holi)
```

```python
df = df.merge(holi, how = 'left', left_on = 'base_ymd', right_on = 'date')
```

```python
def holi_yn(row):
    if row['wn_yn'] =='Y' or row['1'] =='Y'  :
        return 'Y'
    else :
        return 'N'
```

```python
df['holi_yn'] = df.apply(lambda x : holi_yn(x), axis = 1)
```

```python
df.drop(['0','1','2','date'], axis = 1, inplace = True)
```

```python
df['month'] = df.base_ymd.dt.month
```

```python
pv_df = df.groupby(['month','holi_yn','adstrd_nm','tmzon_se_code'], as_index = False).agg({'popltn_cascnt' :'mean'})\
          .pivot_table(index = 'adstrd_nm', columns = ['holi_yn','month','tmzon_se_code'], values = 'popltn_cascnt', fill_value=0)
```

### PCA 주성분분석

```python
# scaling and PCA 
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
data_scale = pd.DataFrame(scaler.fit_transform(pv_df), columns = pv_df.columns, index = pv_df.index)
```

```python
data_scale.head()
```

```python
data_scale = data_scale.loc[:,(slice(None),slice(None),slice('08','22'))]
```

```python
def change_col(i):
    return str(i[0]) +'_'+ str(i[1]) +'M'+ str(i[2])
```

```python
data_scale.columns = list(pd.Series(data_scale.columns.values).apply(change_col))
```

```python
data_scale
```

```python
# PCA 주성분분석
pca = PCA(random_state=1107)
X_p = pca.fit_transform(data_scale)
```

```python
import numpy as np
```

```python
np.cumsum(pca.explained_variance_ratio_)[:5]
```

```python
percent_variance = np.round(pca.explained_variance_ratio_* 100, decimals =2)
columns = []
for i in range(len(percent_variance)):
    columns.append(f'PC{i+1}')
    
plt.bar(x= range(len(percent_variance)), height=percent_variance, tick_label=columns)
plt.ylabel('Percentate of Variance Explained')
plt.xlabel('Principal Component')
plt.xticks(fontsize = 10, color = 'w')
plt.title('PCA Scree Plot')
plt.figure(figsize=(5, 1))
plt.show()
```

```python
pd.DataFrame(pca.components_, columns=data_scale.columns, index=columns).head()
```

```python
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(X_pp.values)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Applying K-Means to dataset
kmeans = KMeans(n_clusters=3,init='k-means++',max_iter=100,n_init=10,random_state=0)
y_kmeans = kmeans.fit_predict(X_pp.values)
```

```python
#Applying DBSCAN to dataset
import seaborn as sns
from sklearn.metrics import silhouette_score

print("Seaborn version : ", sns.__version__)
sns.set()
sns.set_style('darkgrid')
sns.set_color_codes()

import warnings
warnings.filterwarnings(action='ignore')
```

```python
# 실루엣 계수 0.5 이상이면 잘 구분된것이다. 
for eps in np.linspace(0.2,1.1,9, endpoint=False):
    _ , axes = plt.subplots(2,4, figsize=(8*4,8), sharex=True)
    for n_s, ax in zip(range(10, 100, 10), axes.flat):
        dbscan = DBSCAN(eps=eps, min_samples= n_s)
        y_kmeans = dbscan.fit_predict(X_pp.iloc[:,:2].values)
        X_pp['k_mean'] = y_kmeans
        if len(np.unique(y_kmeans)) != 1:
            sil_score = silhouette_score(X_pp.iloc[:,:2].values, y_kmeans)
        else :
            sil_score = 0       
        sns.scatterplot(x="PC1", y="PC2", hue = 'k_mean', ax=ax, data=X_pp, legend = 'brief')
        ax.set(title = "eps:{}, n_sample:{}, silhouette:{:.2f}".format(eps, n_s, sil_score))
    plt.show()
```

```python
# 선택 "eps:0.3, n_sample:10"
dbscan = DBSCAN(eps= 0.8, min_samples= 70)
y_kmeans = dbscan.fit_predict(X_pp.iloc[:,:2].values)
X_pp['k_mean'] = y_kmeans
sns.scatterplot(x="PC1", y="PC2", hue = 'k_mean', data=X_pp, legend = 'brief')
plt.title("eps:0.3, n_sample:10")
```

```python
dbscan.fit(X_pp.iloc[:,:2].values)
cluster = dbscan.fit_predict(X_pp.iloc[:,:2].values)
score = silhouette_score(X_pp.iloc[:,:2].values, cluster)
print("실루엣 점수:" , score)
```
